```python
df = spark.read.csv('Pokemon2.csv',inferSchema='True')
```

```python
df = df.withColumnRenamed('_c0','#')
df = df.withColumnRenamed('_c1','Name')
df = df.withColumnRenamed('_c2','Type_1')
df = df.withColumnRenamed('_c3','Type_2')
df = df.withColumnRenamed('_c4','Total')
df = df.withColumnRenamed('_c5','HP')
df = df.withColumnRenamed('_c6','Attack')
df = df.withColumnRenamed('_c7','Defense')
df = df.withColumnRenamed('_c8','Sp.Atk')
df = df.withColumnRenamed('_c9','Sp.Def')
df = df.withColumnRenamed('_c10','Speed')
df = df.withColumnRenamed('_c11','Generation')
df = df.withColumnRenamed('_c12','Legendary')
from pyspark.sql.types import IntegerType
df = df.withColumn("HP",df["HP"].cast(IntegerType()))

df.printSchema()
```
