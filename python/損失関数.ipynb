```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
x = np.linspace(-10,10,1000)
y = x**2
plt.plot(x,y)
```

```python
plt.figure(figsize=(12,8))
delta = 1

a = np.linspace(-delta,+delta,1000)
y = (1/2)*a**2
plt.plot(a,y,c="blue")

a = np.linspace(-delta-5,-delta,1000)
y = delta * (abs(a)-((1/2)*delta))

plt.plot(a,y,c='green')
a = np.linspace(delta,delta+5,1000)
y = delta * (abs(a)-((1/2)*delta))
plt.plot(a,y,c='green')
plt.title(f'delta = {delta}')

```

```python
plt.figure(figsize=(12,8))
delta = 0.05

a = np.linspace(-delta,+delta,1000)
y = (1/2)*a**2
plt.plot(a,y,c="blue")

a = np.linspace(-delta-50,-delta,1000)
y = delta * (abs(a)-((1/2)*delta))

plt.plot(a,y,c='green')
a = np.linspace(delta,delta+50,1000)
y = delta * (abs(a)-((1/2)*delta))
plt.plot(a,y,c='green')
plt.title(f'delta = {delta}')

```
