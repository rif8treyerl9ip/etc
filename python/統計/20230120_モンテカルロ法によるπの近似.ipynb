```python
import numpy as np
import matplotlib.pyplot as plt
```

### モンテカルロ法によるπの近似
引用者様：【M-H法】Python で MCMC を書いてみた
https://fisproject.jp/2015/12/mcmc-in-python/

```python

N = 100000

x = np.random.uniform(-1.0, 1.0, N)
y = np.random.uniform(-1.0, 1.0, N)
```

```python

inside_circle = []

for i in range(N):
    if (x[i]**2 + y[i]**2) < 1.0:
        inside_circle.append(True)
    else:
        inside_circle.append(False)

print(2.0**2 * float(inside_circle.count(True)) / float(N))

plt.scatter(x, y, c=inside_circle, s=5, edgecolor='None')
plt.xlim(-1.0, 1.0)
plt.ylim(-1.0, 1.0)
plt.xlabel('x')
plt.ylabel('y')
plt.show()

```
